```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_excel('/content/drive/MyDrive/wind_main.xlsx')
data.head()

```

```
data['Date']=pd.to_datetime(data.Date)
data
```

```
data['WS'] = data['WS'].astype(float) 
data['WD'] = data['WD'].astype(int) 
data['BP'] = data['BP'].astype(int) 
data['RH'] = data['RH'].astype(int) 
data.dtypes
  
```

```
data.info()
```

```
data.set_index('Date')
data= data.applymap(lambda x: np.NaN if x == 0 else x)
data
```
